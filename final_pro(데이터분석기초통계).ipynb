```python
main_data['orgNm'] = main_data['orgNm'].str.split(' ',expand = True)[1]
main_data
```

```python
main_data_process = main_data[['orgNm', 'processState']]
main_data_process['processState'].unique()
```

```python
main_data_prepro = main_data_process[main_data_process['processState']=='기증'].groupby(['orgNm']).count()
main_data_prepro = main_data_prepro.iloc[:, 0].to_frame().reset_index()
main_data_prepro.rename(columns={"orgNm":"자치구명", "processState":"기증"}, inplace=True)
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='입양'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['입양'] = m['processState']
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='자연사'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['자연사'] = m['processState']
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='안락사'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['안락사'] = m['processState']
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='반환'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['반환'] = m['processState']
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='보호중'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['보호중'] = m['processState']
main_data_prepro
```

```python
m = main_data_process[main_data_process['processState']=='방사'].groupby(['orgNm']).count()
m = m.iloc[:, 0].to_frame().reset_index()
main_data_prepro['방사'] = m['processState']
main_data_prepro
```

```python
main_data_prepro.fillna(0, inplace=True)
```

```python
main_data_prepro[['기증','입양','반환','자연사','안락사','방사','보호중']] = main_data_prepro[['기증','입양','반환','자연사','안락사','방사','보호중']].astype('int32')
main_data_prepro.info()
```

```python
columns_to_convert = ['기증', '입양', '자연사', '안락사', '반환', '보호중', '방사']

for column in columns_to_convert:
    main_data_prepro[column] = (main_data_prepro[column] / main_data_prepro['발생건수']) * 100

main_data_prepro
```
